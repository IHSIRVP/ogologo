```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_random_images_from_directory(directory_path, num_images=5):
    # Get a list of files in the directory
    files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
    
    # Select a random subset of files
    selected_files = random.sample(files, min(num_images, len(files)))
    
    # Display the images
    plt.figure(figsize=(20, 10))
    for i, file_path in enumerate(selected_files, start=1):
        img = mpimg.imread(file_path)
        plt.subplot(1, len(selected_files), i)
        plt.imshow(img)
        plt.axis('off')
    plt.show()

# Set the path to the directory containing the images
directory_path = "C:/Users/meena/Downloads/LLD-logo_sample"

# Display 5 random images from the directory
display_random_images_from_directory(directory_path, num_images=9)
```

```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_random_images_from_directory(directory_path, num_images=9):
    # Ensure num_images is not greater than 9 for a 3x3 grid
    num_images = min(num_images, 9)
    
    # Get a list of files in the directory
    files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
    
    # Select a random subset of files
    selected_files = random.sample(files, min(num_images, len(files)))
    
    # Display the images
    plt.figure(figsize=(20, 20))  # Adjust the figure size as needed
    for i, file_path in enumerate(selected_files, start=1):
        img = mpimg.imread(file_path)
        plt.subplot(3, 3, i)  # Change to 3x3 grid
        plt.imshow(img)
        plt.axis('off')
    plt.show()

# Set the path to the directory containing the images
directory_path = "C:/Users/meena/Downloads/LLD-logo_sample"

# Display images in a 3x3 format from the directory
display_random_images_from_directory(directory_path)
```
